## import the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## Data Collection

```python
# Loading the dataset to a pandas datadframe
credit_card_data = pd.read_csv('data.csv')
```

## This Data set is highly unbalanced 
0-> Normal Transaction
1-> Fraudelent Transaction

```python
## Separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape,fraud.shape)
```

```python
fraud.Amount.describe()
```

## Under-Sampling

### Build a Sample data set containing the similar distribution of normal transaction and Fraudelent Transaction

#### Number of Fraudelent Transaction ->492
#### Number of Normal Transaction->492

```python
legit_sample = legit.sample(n=492)
```

## Concatenating the two DataFrame

```python
new_dataset  = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting the data into Features and Labels

```python
X = new_dataset.drop(['Class'],axis=1)
Y = new_dataset['Class']
```

```python
print(X.shape,Y.shape)
```

```python
print(X,Y)
```

## Split the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training 

Using Logistic Regression because it is Binary Classification Problem

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
```

## Model Evaluation

Using Accuracy Score

```python
# Accuracy On Training Data
X_train_prediction = model.predict(X_train)
Accuracy_of_training_data = accuracy_score(X_train_prediction,Y_train)
print('Accuracy of training data: ',Accuracy_of_training_data)
```

```python
# Accuracy On Test Data
X_test_prediction = model.predict(X_test)
Accuracy_of_testing_data = accuracy_score(X_test_prediction,Y_test)
print('Accuracy of testing data: ',Accuracy_of_testing_data)
```
